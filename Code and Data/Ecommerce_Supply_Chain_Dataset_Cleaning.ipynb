# **Ecommerce Order And Supply Chain Dataset Preparation to use for Analysis.**

1. Importing pandas to convert file into a pandas Dataframe.

```python
import pandas as pd
```

2. Converting and merging the two csv files into one pandas Dataframe.

```python
path_to_csv = '/df_Customers.csv'
path_to_csv1 = '/df_Orders.csv'
csv = pd.read_csv(path_to_csv, encoding="latin-1")
csv1 = pd.read_csv(path_to_csv1, encoding="latin-1")
df = pd.DataFrame(csv)
df1 = pd.DataFrame(csv1)
merged_df = pd.merge(df,df1, on = 'customer_id')
merged_df
```

3. Checking for any missing values in our Dataframe using IsNull

```python
print(pd.isnull(merged_df).sum())
```

4. With null values present in the order_delivered_timestamp column and the order_approved_at column, I want to make sure these aren't completed orders. If they are orders that are still being processed, I will replace the null values in order_delivered_timestamp with "pending".

```python
processing_orders = merged_df[merged_df['order_status'] == 'processing']
processing_orders
```

5. Here we are assigning the result of the '.fillna' back to the original dataframe to avoid a FutureWarning that happens when modifying a dataframe in place.

```python
merged_df['order_delivered_timestamp'] = merged_df['order_delivered_timestamp'].fillna('pending')
merged_df['order_approved_at'] = merged_df['order_approved_at'].fillna('unknown')
print(pd.isnull(merged_df).sum())
```

6. With null values out of the way, I want to make sure there are no duplicates in this Dataframe.

```python
print(merged_df.duplicated().sum())
```

7. Before we export our database to a CSV, I want to make sure all of the columns are the correct data type.

```python
print(merged_df.dtypes)
```

8. Our dates in this database are object data types. Before export, I need to convert them to datetime format.

```python
# Here we strip the strings to make for conversion to datetime not return any errors.
merged_df['order_purchase_timestamp'] = merged_df['order_purchase_timestamp'].str.strip()
merged_df['order_approved_at'] = merged_df['order_approved_at'].str.strip()
merged_df['order_delivered_timestamp'] = merged_df['order_delivered_timestamp'].str.strip()
merged_df['order_estimated_delivery_date'] = merged_df['order_estimated_delivery_date'].str.strip()

# Since we replaced null values earlier with strings, here we are replacing them with Not a time values
# to ensure no error when converting.
merged_df['order_approved_at'] = merged_df['order_approved_at'].replace(
    to_replace=r'unknown|invalid_date',
    value=pd.NaT,
    regex=True
)

merged_df['order_delivered_timestamp'] = merged_df['order_delivered_timestamp'].replace(
    to_replace=r'pending|invalid_date',
    value=pd.NaT,
    regex=True
)

# Conversion process of the 4 columns.
merged_df['order_purchase_timestamp'] = pd.to_datetime(merged_df['order_purchase_timestamp'])
merged_df['order_approved_at'] = pd.to_datetime(merged_df['order_approved_at'])
merged_df['order_delivered_timestamp'] = pd.to_datetime(merged_df['order_delivered_timestamp'])
merged_df['order_estimated_delivery_date'] = pd.to_datetime(merged_df['order_estimated_delivery_date'])
merged_df
print(merged_df.dtypes)
```

9. Now that missing values have been fixed and data has been formatted correctly. I will convert this database into a CSV to be used in an SQL editor.

```python
merged_df.to_csv('/EcommerceDatasetCleaned.csv')
```
